```python
df1 = pd.read_csv("train1.csv")
df2 = pd.read_csv("train2.csv")

df1 = df1.iloc[:100, 1:-2]
df2 = df2.iloc[:, 1:-2]
df = pd.concat([df1, df2], axis = 0)
```

```python
#건입 ver.OG

X_1, X_2 = df[["e_longitude", "d_longitude", "o_longitude"]], df[["e_latitude", "d_latitude", "o_latitude"]]
y_1, y_2 = df["s_longitude"], df["s_latitude"]


test1 = pd.DataFrame({"e_longitude" : [922], "d_longitude" : [911], "o_longitude" : [1002]})
test2 = pd.DataFrame({"e_latitude" : [463], "d_latitude" : [460], "o_latitude" : [470]})

S = [[1040, 602], [1032, 420], [1120, 594]]


Model = [LinearRegression(), svm.SVR(), XGBRegressor(), LGBMRegressor(), RandomForestRegressor(), GradientBoostingRegressor()]
temp1, temp2 = [], []
for i in range(len(Model)):
    for j in range(len(Model)):
        if i == j:
            continue
        else:
            model1, model2 = Model[i], Model[j]
            model1.fit(X_1, y_1)
            model2.fit(X_2, y_2)
            temp1.append(model1.predict(test1)[0])
            temp2.append(model2.predict(test2)[0])
res = pd.DataFrame({"x" : temp1, "y" : temp2})
res["D"] = 100000000
res["idx"] = -1
for i in range(len(res)):
    for s in S:
        temp = ((s[0] - res["x"][i])**2 + (s[1] - res["y"][i])**2)**0.5
        if res["D"][i] >= temp:
            res["D"][i] = temp
            res["idx"][i] = S.index(s)
res.sort_values("D")
```
